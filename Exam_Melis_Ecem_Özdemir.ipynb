```python
##Q1 calculating moving average
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import random as rn
L = 200
list1= []
for x in range(0,L):
    list1.append(rn.randint(1,100))

mov_avg = []

for i in range(1, L):
    avg = 0
    for j in range(i, 0, -1):
        avg = avg + list1[j-1]

    mov_avg.append(round(avg/L,4))

print(list1)

print(mov_avg)
plt.plot(mov_avg)
plt.plot(list1)
plt.show()
```

```python
#Q2.1european call option
import numpy as np
import random

tbegin = 0
tend = 1
deltat = .0001
t = np.arange(tbegin, tend, deltat)

K=1.0

sigma = 0.3
S0 = 1

sqrtdt = np.sqrt(deltat)

N=np.zeros(1000) 
for i in range(0,N.size):
    y = S0
    for j in range(1, t.size):
        y = y * np.exp(random.choice([-1,1]) * sigma * sqrtdt)
    N[i]=y

N=np.maximum(0,N-K)
print(N.mean())
```

```python
#Q2.2european put option
import numpy as np
import random

tbegin = 0
tend = 1
deltat = .0001
t = np.arange(tbegin, tend, deltat)

K=1.0

sigma = 0.3
S0 = 1

sqrtdt = np.sqrt(deltat)

N=np.zeros(1000) 
for i in range(0,N.size):
    y = S0
    for j in range(1, t.size):
        y = y * np.exp(random.choice([-1,1]) * sigma * sqrtdt)
    N[i]=y

N=np.maximum(0,K-N)
print(N.mean())
```
